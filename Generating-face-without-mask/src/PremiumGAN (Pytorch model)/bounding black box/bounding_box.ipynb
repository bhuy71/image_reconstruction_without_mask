```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Load the trained model
model = YOLO('C:\\Sam C-Gan project\\bounding black box\\mask_detection.pt')

# Perform inference on a new image or directory of images
results = model.predict(source='C:\\Sam C-Gan project\\bounding_box_data\\val\\images\\with-mask-default-mask-seed0452.png', save=False, conf=0.1)

# Loop through results and display each image
for result in results:
    # Get the image with the predicted boxes and labels drawn on it
    img = result.plot()  # Use .plot() to get the image with predictions drawn

    # Convert BGR image (from OpenCV) to RGB for display with matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Display the image
    plt.imshow(img_rgb)
    plt.axis('off')  # Hide the axes
    plt.show()
```

```python
import cv2
import numpy as np
from ultralytics import YOLO

def convert(image_path, model):
    img = cv2.imread(image_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    results = model.predict(source=img, save=False)

    if results:
        boxes = results[0].boxes

        for box in boxes:
            xyxy = box.xyxy[0].cpu().numpy()
            x1, y1, x2, y2 = map(int, xyxy)

            cv2.rectangle(img_rgb, (x1, y1), (x2, y2), color=(0, 0, 0), thickness=-1)

    return img_rgb

model = YOLO('C:\\Sam C-Gan project\\bounding black box\\mask_detection.pt')
output_image = convert('C:\\Sam C-Gan project\\bounding_box_data\\val\\images\\with-mask-default-mask-seed0452.png', model)

plt.imshow(output_image)
plt.axis('off')
plt.show()

```
